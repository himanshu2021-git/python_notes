## Pie Chart

these charts are commonly used to visualize the distributions

```python
covid_details_india = [494657, 8013783, 127571]
slices_name = ["Active Cases","Recovered Cases","Fatal Cases"]
colors = list('bgr')

plt.pie(covid_details_india,labels = slices_name,colors=colors)
plt.title('Covid-19 Spread Over India')
```

```python
plt.pie(
    covid_details_india,
    labels = slices_name,
    colors=colors,
    startangle=180,
    shadow=True,
    explode = (0.1,0.1,0.5),
    autopct = "%.2f%%"
)
plt.title('Covid-19 Spread Over India')
```

## Stacked Bar Chart

These are generally a good way to represent multiple types of data within a single bar.

```python
cities = ['New Delhi','Gurugram',"Noida","Gaziabaad","Faridabaad"]
men_age_mean = [45,35,79,74,35]
women_age_mean = [65,25,9,24,5]
```

```python
width = 0.5

fig,ax = plt.subplots()

ax.bar(
    cities,
    men_age_mean,
    width,
    label = "Men"
)

ax.bar(
    cities,
    women_age_mean,
    width,
    bottom = men_age_mean,
    label="Women"
)

ax.set_ylabel('Mean Ages')
ax.set_title("Average age of Men and Women over NCR")
ax.legend()
```

```python
x = np.arange(len(cities))

width = 0.35

fig,ax = plt.subplots()

c1 = ax.bar(x - width/2,men_age_mean,width,label="Men")
c2 = ax.bar(x + width/2,women_age_mean,width,label="Women")

ax.set_ylabel("Mean Ages")
ax.set_title("Average age of Men and Women over NCR")
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()
```

## Horizontal Bar Charts

This chart is best for visualizing the <b> Nominal, Ordinal or Interval </b> Data 

```python
apps = "Facebook","Tiktok","Insta","WhatsApp","Twitter"
y = np.arange(len(apps))
downloads = (45,50,65,75,35)

fig,ax = plt.subplots()

ax.barh(y,downloads,align="center")
ax.set_yticks(y)
ax.set_yticklabels(apps)
ax.invert_yaxis()
ax.set_xlabel("Downloads in Millions")
ax.set_title('Best App Over Play Store')
```

```python
activity = ["Gaming",'Driving','Biking','Music Listening']
males = ["Love","Love","Love","Hate"]
females = ["Hate","Love","Hate","Love"]

fig,ax = plt.subplots()

ax.plot(activity,males,label="Male")
ax.plot(activity,females,label="Female")
ax.legend()
```

```python
apps_rating = {'Insta':4,'WhatsApp':5,'Twitter':3.5,'Facebook':4.2}

fig,ax = plt.subplots(1,3,figsize=(9,3),sharey=True)

x,y = list(apps_rating.keys()),list(apps_rating.values())

ax[0].bar(x,y)
ax[1].scatter(x,y)
ax[2].plot(x,y)
ax[0].tick_params(labelrotation=45)
ax[1].tick_params(labelrotation=45)
ax[2].tick_params(labelrotation=45)
ax[0].set_title('BAS')

fig.suptitle("Apps Rating")
```

## Multiple Plots

```python
fig,axs = plt.subplots(1,2) # if you provide only a single value that will represent the no of rows
x = np.linspace(0,2*np.pi,400)
y = np.sin(x ** 2)
axs[0].plot(x,y)
axs[1].plot(x,y)
fig.suptitle('VERTICALLY STACKED SUBPLOTS')
```

```python
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title('C-0 0')
axs[0,1].plot(x,y)
axs[0,1].set_title('C-0 1')
axs[1,0].plot(x,y)
axs[1,0].set_title('C-1 0')
axs[1,1].plot(x,y)
axs[1,1].set_title('C-1 1')
#adding the x and y labels
for ax in axs.flat:
    ax.set(xlabel="x-label", ylabel="y-label")
# remoing the inner ticks
for ax in axs.flat:
    ax.label_outer()
```

## Customizing the Lines

```python
fig,ax = plt.subplots()
x = np.linspace(0,10,500)
y = np.sin(x)
line1 = ax.plot(x,y,label='Custom lines',dashes=[20,2,10,2])
```

```python
x = np.linspace(0.1,2*np.pi,41)
y = np.exp(np.sin(x))

plt.stem(x,y,use_line_collection=True)
```

```python
fig,ax = plt.subplots()

t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)

line, = ax.plot(t,s,lw=2)

ax.annotate('my notation',xy=(1,0.5),xytext=(0.75,3),arrowprops=dict(facecolor="black",shrink=0.05))

ax.set_ylim(-2,2)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="smoker",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="time",data=tips)
```

```python
tips.day.unique()
```

```python
sns.catplot(x="day",y="total_bill",data=tips)
```

```python
# who tips more smoker or non smoker

sns.catplot(x="smoker",y="tip",data=tips)
```

```python
# which day of week has the biggest total_bill amount on dinner or lunch

sns.catplot(x="total_bill",y="day",hue="time",data=tips)
```
