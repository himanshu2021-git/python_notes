```python
from random import choice,randint
data = {
    'Name':['KRISHNA','rama','RAHUL','MANOJ'],
    'Nationality':[choice(['indian','american']) for x in range(4)],
    'gender':list('MMMM'),
    'age':[randint(20,40) for x in range(4)],
    'nature':['Polite','Angry','Polite','Angry']
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
pd.crosstab(df.Nationality,df.nature)
```

```python
pd.crosstab(df.Nationality,df.nature,margins=True)
```

```python
pd.crosstab([df.Nationality,df.gender],[df.nature,df.age],margins=True)
```
