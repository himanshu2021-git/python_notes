### How to create pandas series object
in real world the pandas series are taken from some resource(file,web,database,etc)


```python
#blank pandas series
s = pd.Series()
s
```

```python
arr = np.array(list("krishna"))
arr
```

```python
#creating pandas series from numpy array
s = pd.Series(arr) # if you do not provide the label then numerical label will be used automatically.
s
```

```python
#creating pandas series from dictionary of python
data = {chr(x+97):x for x in range(26)}
list(data.items())[:10]
```

```python
s = pd.Series(data)
s
```

```python
#providing the indexes explicitly.
s = pd.Series(list("krishna"),index=range(10,17))
s
```

```python
#creating series from scaler values
s = pd.Series(20,index=range(20))
s
```
