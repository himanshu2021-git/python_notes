```python
df = pd.read_csv("https://bit.ly/smallstocks")
```

```python
df
```

```python
# DIRECT ITERATION OVER A DATAFRAME WILL ALWAYS GETS THE COLUMN NAMES ONLY
for x in df:print(x)
```

```python
# this will allow you to access column name and data of it
for column_name,data in df.iteritems():print(column_name,data,sep="\n\n",end="\n"+("-"*10))
```

```python
# iterating over the rows
for label,data in df.iterrows():print(label,data,sep="\n\n")
```

```python
for x in df.itertuples():print(x)
```

```python
filter_symbol = {
    'CSCO':'CISCO',
    'AAPL':'APPLE',
    'MSFT':'MICROSOFT'
}
def filter_sym(cell):
    return filter_symbol.get(cell,'DEFAULT')
```

```python
df = pd.read_csv('https://bit.ly/smallstocks',converters = {'Symbol':filter_sym})
```

```python
df
```
