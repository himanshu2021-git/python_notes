# CONCATENATE METHOD OF PANDAS
this method can be used to merge the two or more dataframes on some axises

```python
df1 = pd.DataFrame(
    {
        'age':range(10,20),
        "name":list("ABCDEFGHIJ")
    }
)
```

```python
df2 = pd.DataFrame(
    {
        'age':range(10,20),
        "name":list("KLMNOPQRST")
    }
)
```

```python
df1
```

```python
df2
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df3 = pd.concat([df1,df2],ignore_index=True)
df3
```

```python
df3 = pd.concat([df1,df2],axis=1)
df3
```

```python
df3 = pd.concat([df1,df2],keys = ["Male","Female"],axis=1)
df3
```

```python
df3["Male"]
```

# Crosstab

```python
df = pd.read_csv('natureofhuman.csv')
```

```python
df
```

```python
pd.crosstab(df.Nationality,df.Nature)
```

```python
pd.crosstab(df.Gender,df.Nature,margins=True)
```

```python
pd.crosstab([df.Nationality,df.Gender],[df.Nature,df.Age],margins=True)
```

```python
temp = pd.crosstab([df.Nationality,df.Gender],[df.Nature,df.Age],normalize="index")
```
